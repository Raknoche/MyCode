We could try to predict fantasy points, or try to predict stat outcomes themselves.  I think predicting stats has more value, since the result can be adapted to any fantasy scoring method people come up with.   It also has the added benefit of realizing when a particular defense is strong against running, or passing, or both rather than just strong in general.  The downside is that it may be harder to predict individual stats than it would be to predict scores.

Let's give predicting stats a shot and see how it works out.  As a first attempt, we should just try to predict something like rushing yards.  First, we need to produce a list of samples, including features and their outcome.  The samples will be stats for an individual player for an individual game.  We'll scan all games played one at a time to collect this information.  What should the features we use for our prediction be?


To start, we can try these features:

* DONE: Average relevant stats for the player (From the last X games prior to a game)
* DONE: Player's team
* DONE: Opponent team
* DONE: Team at home or away
* DONE: Player's average stat against this oppenent
* DONE: Team's average stat against this oppenent (To help account for players switching teams)
* DONE: Player's position (QB's don't rush much, for instance.  This is covered by average stat to begin with)

Here are some more stats we might add in the future.  These may be harder to implement but are worth looking into:

* Player's injury status
* Player's bench status
* DONE: (no coach) Player's coach, other players on his team
* DONE: (no coach) Oppenent's coach, and other players on opponent team

Let's loop through each game and produce our samples for the stats we mentioned in the first list. We'll need to write some functions to calculate the average of the players stats.  **Keep in mind we need to account for unexpected injuries and by weeks... probably just don't include them in the samples at first!**

```python
def calc_avg_for_player(player,year,week):
    '''Calculates average of each stat over the past num_weeks games'''
    #Initializing empty dictionary
    average_stats={
        'passing_yds': 0,
        'passing_tds': 0,
        'passing_twoptm': 0,
        'passing_int': 0,
        'rushing_yds': 0,
        'rushing_tds': 0,
        'rushing_twoptm': 0,
        'receiving_yds': 0,
        'receiving_rec': 0,
        'receiving_tds': 0,
        'receiving_twoptm': 0,
        'kickret_tds': 0,
        'puntret_tds': 0,
        'fumbles_rec_tds': 0,
        'fumbles_lost': 0,
        'passing_att': 0,
        'passing_cmp': 0,
        'passing_cmp_air_yds': 0,
        'passing_sk': 0,
        'passing_sk_yds': 0,
        'passing_twopta': 0,
        'receiving_tar': 0,
        'receiving_twopta': 0,
        'receiving_yac_yds': 0,
        'rushing_att': 0,
        'rushing_twopta': 0}
    
    #number of weeks to average over
    num_weeks=17

    #Number of reg season games
    num_season_games=17

    #The week to start averaging at
    start_week=week-num_weeks
    start_year=year
    end_week=week-1
    end_year=year

    #If start_week is zero or less, we need to go one year back
    #Right now this only handles num_weeks <= 17... would need to change for mutliple years later
    if start_week <= 0:
        start_week = num_season_games+start_week
        start_year = year-1

        #If we step back one week, and end up at zero.. last game is the end of last season
        if end_week == 0:
            end_week = num_season_games
            end_year = year-1

    #initializing averages
    for stat in average_stats:
        exec('total_%s = 0' % stat)
    week_count=0
     
    #If all the stats come from the same year of averaging
    if start_year == end_year:
        for week_num in range(start_week,end_week+1):
            cur_query= nfldb.Query(db)
            cur_query.game(season_year=start_year, season_type='Regular', week=week_num)
            cur_query.player(full_name=player)
            for info in cur_query.as_aggregate():
                #If info is empty, they didn't play that game
                if info:
                    #If it isn't empty, record the stats
                    for stat in average_stats:
                        if stat in dir(info):                                    
                            exec('total_%s += info.%s' % (stat,stat))
                            week_count+=1                        
        
        #After collecting the data, average the stats over the number of weeks
        for stat in average_stats:
            if eval('week_count_%s' % stat)> 0:
                average_stats[stat] = eval('total_%s/week_count' % (stat,stat))
            else:
                average_stats[stat] = None
                
    #If we have to get some stats fromthe previous year
    if start_year != end_year:
        #send a query for the previous year
        for week_num in range (start_week, num_season_games+1):
            prev_query = nfldb.Query(db)
            prev_query.game(season_year=start_year, season_type='Regular', week=week_num)
            prev_query.player(full_name=player)        
            for info in prev_query.as_aggregate():
                if info:
                    for stat in average_stats:
                        if stat in dir(info):                                 
                            exec('total_%s += info.%s' % (stat,stat))
                            week_count+=1                        

        #send a query for the current year
        for week_num in range(1, end_week+1):
            cur_query= nfldb.Query(db)
            cur_query.game(season_year=end_year, season_type='Regular', week=week_num)
            cur_query.player(full_name=player)        
            for info in cur_query.as_aggregate():
                if info:
                    for stat in average_stats:
                        if stat in dir(info):               
                            exec('total_%s += info.%s' % (stat,stat))
                            week_count+=1                        

        for stat in average_stats:
            if eval('week_count_%s' % stat)> 0:
                average_stats[stat] = eval('total_%s/week_count_%s' % (stat,stat))
            else:
                average_stats[stat] = None                            

    return average_stats 
```

```python
def calc_result_for_player(player,year,week):
    '''Calculates average of each stat over the past num_weeks games'''
    #Initializing empty dictionary
    this_game_stats={
        'passing_yds': 0,
        'passing_tds': 0,
        'passing_twoptm': 0,
        'passing_int': 0,
        'rushing_yds': 0,
        'rushing_tds': 0,
        'rushing_twoptm': 0,
        'receiving_yds': 0,
        'receiving_rec': 0,
        'receiving_tds': 0,
        'receiving_twoptm': 0,
        'kickret_tds': 0,
        'puntret_tds': 0,
        'fumbles_rec_tds': 0,
        'fumbles_lost': 0,
        'passing_att': 0,
        'passing_cmp': 0,
        'passing_cmp_air_yds': 0,
        'passing_sk': 0,
        'passing_sk_yds': 0,
        'passing_twopta': 0,
        'receiving_tar': 0,
        'receiving_twopta': 0,
        'receiving_yac_yds': 0,
        'rushing_att': 0,
        'rushing_twopta': 0}
    
    query= nfldb.Query(db)
    query.game(season_year=year, season_type='Regular', week=week)
    query.player(full_name=player)
    for info in query.as_aggregate():
        #If there is no info, they didn't play
        if info:
            for stat in this_game_stats:
                if stat in dir(info):
                    this_game_stats[stat] = eval('info.%s' % stat)
        else:
            for stat in this_game_stats:
                this_game_stats[stat] = None      
         
    return this_game_stats   
```

```python
def getTeamInfo(player,year,week):
    '''Determines the information for a players team'''
    query= nfldb.Query(db)
    query.game(season_year=year, season_type='Regular', week=week)
    query.player(full_name=player)
    for info in query.as_play_players():
        team = info.team
        
    for info in query.as_games():
        team_is_home = (team == info.home_team)
        
        if team_is_home:
            opponent = info.away_team
        else:
            opponent = info.home_team
            
    team_info=[team,team_is_home,opponent]
    
    return team_info
```

```python
#Filling out a dataframe for stats about each game
#Can access stats like so: data['Player Avg Stats'].ix[sample_num]['passing_yds']
data=pd.DataFrame(columns=('Player','Team','At Home','Opponent','Player Avg Stats','Stat Outcomes'))
current_row=0

#Make a list of all players
allGames = nfldb.Query(db)
allGames.game(season_year=2015, season_type='Regular')
all_players=[]
player_positions=[]
player_teams=[]
for info in allGames.as_aggregate():
    if str(info.player).split(' (')[0] not in all_players:
        all_players.append(str(info.player).split(' (')[0])
        player_positions.append(str(info.player).split(' (')[1].split(', ')[1].split(')')[0])
        player_teams.append(str(info.player).split(' (')[1].split(',')[0])
        
for week_num in range(1,2):
    for player in all_players:
        avg_stats=calc_avg_for_player(player,2015,week_num)
        stat_outcomes = calc_result_for_player(player,2015,week_num)
        team_info = getTeamInfo(player,2015,week_num)
                          
        data.loc[current_row] = [player, team_info[0], team_info[1], team_info[2], avg_stats, stat_outcomes]
        current_row+=1
```

# New Approach

The above code works pretty well... but it is kind of slow to run, since we send a query for every player individually.
```python
def FindPlayerAverage(num_weeks,player,current_week,opp=None):
    '''Averages a players stats over the past num_weeks
    If we dont have num_weeks of data, it goes back as far as possible
    
    Inputs:
    num_weeks -> number of weeks to average over
    player    -> player whose stats we want to average
    current_week -> The corresponding week in gameStats['Week'] for the sample
    opp       -> optional.  Specify opponent to average over games versus that opponent
    '''
    
    start_week = current_week - num_weeks + 1
    playerStats = gameStats[(gameStats['Player']== player) & (gameStats['Week']>= start_week)].reset_index(drop=True)
    
    #Check if all the stats are empty.  If so, assume they didn't play
    playerStats['Played'] = True
    for row in range(len(playerStats)):
        if max(playerStats[key].ix[row] for key in this_game_stats) <= 0:
            playerStats['Played'].ix[row] = False
         
    #Remove games where they player didn't play
    playerStats=playerStats[playerStats['Played']==True]
 
    #Average the remaining stats
    playerAverage=[np.mean(playerStats[key]) for key in this_game_stats]

    if opp:
        playerStats_v_opp = playerStats[playerStats['Opponent']==opp]
        playerAverage_v_opp=[np.mean(playerStats_v_opp[key]) for key in this_game_stats]
        return (playerAverage,playerAverage_v_opp)
        
    else:
        return playerAverage
```

```python
def FindTeamAverage(num_weeks,team,current_week,opp=None):
    '''Averages a teams stats over the past num_weeks
    If we dont have num_weeks of data, it goes back as far as possible
    
    Inputs:
    num_weeks -> number of weeks to average over
    team      -> team whose stats we want to average
    current_week -> The corresponding week in gameStats['Week'] for the sample    
    opp       -> optional.  Specify opponent to average over games versus that opponent
    '''
    start_week = current_week - num_weeks + 1
    teamStats = gameStats[(gameStats['Team']== team) & (gameStats['Week']>= start_week)].reset_index(drop=True)

    #Average the remaining stats
    teamAverage=[np.mean(teamStats[key]) for key in this_game_stats]

    if opp:
        teamStats_v_opp = teamStats[teamStats['Opponent']==opp]
        teamAverage_v_opp=[np.mean(teamStats_v_opp[key]) for key in this_game_stats]
        return (teamAverage,teamAverage_v_opp)
    else:
        return teamAverage
```

Now we need to find the team's average stats against the opposing team, and the player's average stats against the opposing team.  Then, we'll create a list of samples and features. 

```python
import time
```

```python
#Making trainingset with these features
'''
* DONE: Average relevant stats for the player (From the last X games prior to a game)
* DONE: Player's team
* DONE: Opponent team
* DONE: Team at home or away
* DONE: Player's average stat against this oppenent
* DONE: Team's average stat against this oppenent (To help account for players switching teams)
* Player's position (QB's don't rush much, for instance.  This is covered by average stat to begin with)
'''


trainingSamples=pd.DataFrame(columns=('Player','Team','Opponent','At Home','Player Avg Stats','Player Avg v Opp','Team Avg Stats','Team Avg v Opp',))


t0 = time.time()
num_samps=10
for row in range(num_samps):
    trainingSamples.loc[row]=None
    player = gameStats['Player'].ix[row]
    team = gameStats['Team'].ix[row]
    opp = gameStats['Opponent'].ix[row]
    num_weeks=17 #number of weeks to average over
    player_stats, player_stats_v_opp = FindPlayerAverage(num_weeks,player,gameStats['Week'].ix[row],opp)
    team_stats, team_stats_v_opp = FindTeamAverage(num_weeks,team,gameStats['Week'].ix[row],opp)

    trainingSamples['Player'].loc[row] = player
    trainingSamples['Team'].loc[row] = team
    trainingSamples['Opponent'].loc[row] = opp
    trainingSamples['At Home'].loc[row] = gameStats['At Home'].ix[row]
    trainingSamples['Player Avg Stats'].loc[row] = player_stats
    trainingSamples['Player Avg v Opp'].loc[row] = player_stats_v_opp
    trainingSamples['Team Avg Stats'].loc[row] = team_stats
    trainingSamples['Team Avg v Opp'].loc[row] = team_stats_v_opp    
 

t1 = time.time()
total = t1-t0

print "Total Run Time for ", num_samps, " samples is " , total, " seconds."
print "Avering ", total/num_samps, " seconds per sample.  Estimated ful run time is ", len(gameStats)*total/num_samps/60/60, " hours per year of data"
```
