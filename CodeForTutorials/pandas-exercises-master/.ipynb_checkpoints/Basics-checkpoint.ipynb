Series
======

```python
#randn creates a numpy array of random data
np.random.randn(5)
```

```python
#Create labels for the numpy array generated by randn
#and assign it to a 1xN Series
labels = ['a', 'b', 'c', 'd', 'e']
s = Series(np.random.randn(5), index=labels)
s
```

```python
#Series behaves very similar to a dict
#Check if label b is in the series
'b' in s
```

```python
#Get value at label b
s['b']
```

```python
#array of the label indices
s.index
```

```python
#Convert the series to a dict
mapping = s.to_dict()
mapping
```

```python
#Make a series from a dict, an specify the order from the keys
#If a key isn't included in the index list, it isn't put in the series
#If a nonexitent key is included in the index list, it is filled with NaN
s = Series(mapping, index=['b', 'e', 'a', 'd', 'f'])
s
```

```python
#notnull and isnull create boolean numpy arrays
#Can use these for logical cuts on the series where data is/isn't null
s[notnull(s)]

#There is also isnan, which only works on floating point arrays
```

```python
#Shortcut to select all NA values from the series with .dropna
#Note, this doesn't actually change the series
s.dropna()
```

```python
#Can do math with the series like in numpy arrays
#The main difference is you ALSO get the label returned
s * 2
```

```python
#Can do slicing with the series like in numpy
#Get a slice of both the data AND the labels
s[:3]
```

DataFrame: 2D collection of Series
==================================

```python
#Think of a dataframe as a dict of arrays
#This is one way to make a dataframe

df = DataFrame({'a': np.random.randn(6),
                'b': ['foo', 'bar'] * 3,
                'c': np.random.randn(6)})

#Can append columns to the dataframe later with indexing
#Can assign scalar values to the entire column as well df['d']=5
df['d'] = range(6)

#Display the entire data fram
df
```

```python
#Can call a certain column with indexing df['a']
df['b']
```

```python
#Can get rows with slicing
df[:3]
```

```python
#Can make a series from a row, using column names as labels
df.xs(0)   #Here, we get row zero
```

```python
#The .ix attribute allows indexing like a matrix
df.ix[2, 'b']
```

```python
#Alternatively, you can use get_value to get dataframe values
#This is faster than looking up data with .ix
df.get_value(2, 'b')
```

```python
#Can use .ix to get just a row, or just a column too
df.ix[2]
```

```python
#Can include slicing in .ix
df.ix[2:4, 'b']
```

```python
#Can also slice with column labels
#Note 'c':'b' would return nothing since c comes after b in the labels
df.ix[2:4, 'b':'c'] 
```

```python
#Can specify particular rows or columns that you want
df.ix[[0, 2, 4], ['b', 'c', 'd']] 
```

```python
#Can produce logical cuts
df['c'] > 0
```

```python
#Can pass the logical cuts into .ix to .get_value
df.ix[df['c'] > 0] 
```

```python
#Can specify the index when making the data frame
df = DataFrame({'a': np.random.randn(6),
                'b': ['foo', 'bar'] * 3,
                'c': np.random.randn(6)},
               index=date_range('1/1/2000', periods=6))
df
```

```python
#Can specify the column order when making the data frame
#Extra columns will be filled with NaN
df = DataFrame({'a': np.random.randn(6),
                'b': ['foo', 'bar'] * 3,
                'c': np.random.randn(6)},
               columns=['a', 'b', 'c', 'd'])
df
```

Creation from nested dicts
--------------------------

These arise naturally in Python code

```python
#Creating a nested dict for example purposes
data = {}
for col in ['foo', 'bar', 'baz']:
    for row in ['a', 'b', 'c', 'd']:
        data.setdefault(col, {})[row] = np.random.randn()   # sorted cols: bar, baz, foo
data
```

```python
del data ['foo']['c']   # delete an entry by index key (col) and row.
data
```

```python
#Can pass the nested dict into DataFrame to make a df from it
#Outer keys become the columns
#Deleted dict entries will show up as NaN
DataFrame(data)


#Can do this with lists of lists, dicts of lists, etc as well
```

Data alignment
==============

```python
#Load a csv file into python using pandas.read_csv
close_px = read_csv('stock_data.csv', index_col=0, parse_dates=True)
```

```python
#Display the csv data
!head -n 10 stock_data.csv
```

```python
#The pandas df from the csv
close_px
```

```python
#Picking pieces out from the dataframe with slicing
s1 = close_px['AAPL'][-20:]      # the last 20 rows
s2 = close_px['AAPL'][-25:-10]   # the last 25 - last 10 = 15 rows.

#Displaying the data series side by side
#Note sidy_by_side was defined in line one of the notebook
side_by_side(s1, s2)
```

```python
#Sum the two series where we can (similar to MySQL SUM)
#Filled with NaN if we can't join the two series
s1 + s2
```

```python
#Add the two series, but instead of filling NaN when we can't join
#It will fill missing data with zeros 
#(So some of the sums here are just from one series)
s1.add(s2, fill_value=0)
```

```python
#As shown before, we could drop the NaN with .dropna attribute
(s1 + s2).dropna()
```

```python
#Can also make sub-dataframes with slicing
df = close_px.ix[-10:, :3]
df
```

```python
#Can reorder the index of a series with .reindex
#Will fill NaN where we can't join
side_by_side(s1.reindex(s2.index), s2) 
```

```python
#Could also use .ix for this
side_by_side(s1.ix[s2.index], s2)   # same as above, supported in newer version of 0.12 ipython.
```

```python
#Inner joins
b, c  = s1.align(s2, join='inner')
side_by_side(b, c)
```

```python
#Left Joins
b, c  = s1.align(s2, join='outer')
side_by_side(b, c)
```

```python
#Right Joins
b, c  = s1.align(s2, join='right')
side_by_side(b, c)
```

```python
#Can do all of the above with data frames as well
df = close_px.ix[-10:, ['AAPL', 'IBM', 'MSFT']]
df
```

```python
df2 = df.ix[::2, ['IBM', 'MSFT']]
side_by_side(df, df2)
```

```python
df + df2
```

```python
b, c = df.align(df2, join='inner')
side_by_side(b, c) 
```

GroupBy example: linear regression by group
-------------------------------------------

```python
import scikits.statsmodels.api as sm
rets = close_px / close_px.shift(1) - 1

def get_beta(rets):
    rets = rets.dropna()
    rets['intercept'] = 1.
    model = sm.OLS(rets['MSFT'], rets.ix[:, ['AAPL', 'intercept']]).fit()
    return model.params

get_beta(rets)
```

```python
grouped = rets.groupby([lambda x: x.year, lambda x: x.month])
beta_by_ym = grouped.apply(get_beta)
beta_by_ym
```

```python
beta_by_ym.unstack(0)['AAPL']
```

GroupBy with hierarchical indexing
----------------------------------

```python
tuples = zip(*[['bar', 'bar', 'baz', 'baz',
                'foo', 'foo', 'qux', 'qux'],
               ['one', 'two', 'one', 'two',
                'one', 'two', 'one', 'two']])
index = MultiIndex.from_tuples(tuples)
columns = MultiIndex.from_tuples([('A', 'cat'), ('B', 'dog'),
                                  ('B', 'cat'), ('A', 'dog')])
df = DataFrame(randn(8, 4), index=index, columns=columns)
df
```

```python
df.groupby(level=0, axis=0).mean()
```

```python
df.stack()
```

```python
df.stack().mean(1).unstack()
```

```python
# could also have done
df.groupby(level=1, axis=1).mean()
```
